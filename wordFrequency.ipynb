```python
with open('/content/wordfrequency-dataset.txt', 'r') as f:
    read_text = f.read()
```

```python
sentences = re.split('[.,:•()"\n\t]', read_text)

for i in range(len(sentences)):
  sentences[i] = sentences[i].lower()

sentences = ' '.join(sentences)
sentences = sentences.split()
print(sentences)
```

```python
sentences
```

# **Defining the window size**

```python
window = 2
frequency = {}
```

```python
sentences
```

```python
temp = []
for i in range(len(sentences)-window):
  temp.append(sentences[i:i+window])

temp
```

```python
tuple(temp[0])
```

```python
for i in range(len(temp)):
  if tuple(temp[i]) not in frequency:
    frequency[tuple(temp[i])] = 1
  else:
    frequency[tuple(temp[i])] += 1

frequency
```

```python
import operator

sorted_frequency = dict(sorted(frequency.items(), key=operator.itemgetter(1), reverse=True))
sorted_frequency
```

```python
with open('sorted_frequency.pickle', 'wb') as handle:
    pickle.dump(sorted_frequency, handle, protocol=pickle.HIGHEST_PROTOCOL)

from google.colab import files
files.download('sorted_frequency.pickle')
```
